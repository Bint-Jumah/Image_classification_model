```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.datasets import mnist
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory('/content/drive/MyDrive/train',
                                                               image_size=(224,224),
                                                               batch_size=32
                                                               )
```

```python
!ls /content/drive/MyDrive/train
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory('/content/drive/MyDrive/train',
                                                               image_size=(224,224),
                                                               batch_size = 32
)
```

```python
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224,224,3),
    include_top=False,
    weights='imagenet'
)
```

```python
base_model.trainable = False
```

```python
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(train_ds, validation_data= val_ds, epochs=5)
```

```python
model.evaluate(val_ds)
```

```python
from sklearn.metrics import classification_report
```

```python
converter= tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model= converter.convert()

with open('model.tflite', 'wb') as f:
  f.write(tflite_model)
```
